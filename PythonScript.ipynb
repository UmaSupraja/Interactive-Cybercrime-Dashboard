## CYBER CRIMES IN INDIA (SINCE 2002-2020)

```python
### Importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### Load Data Set

df = pd.read_csv("cc.csv")  # ensure you've saved PDF table as CSV first
df
```

```python
# Returns the number of rows and columns in the DataFrame
# Syntax: df.shape
print(df.shape)

# Lists all column names in the DataFrame
# Syntax: df.columns
print(df.columns)

# Displays the data type of each column
# Syntax: df.dtypes
print(df.dtypes)

# Provides a concise summary including types and non-null values
# Syntax: df.info()
print(df.info())

# Gives statistical summary like count, mean, std, etc. for numeric columns
# Syntax: df.describe()
print(df.describe())

# Shows total number of missing (NaN) values in each column
# Syntax: df.isnull().sum()
print(df.isnull().sum())

# Shows the count of unique values in each column
# Syntax: df.nunique()
print(df.nunique())

# Displays the first 5 rows of the DataFrame
# Syntax: df.head()
print(df.head())

# Displays the last 5 rows of the DataFrame
# Syntax: df.tail()
print(df.tail())

# Returns a random sample of n rows from the DataFrame
# Syntax: df.sample(n)
print(df.sample(5))

# Lists all unique values from a specific column
# Syntax: df['ColumnName'].unique()
print(df['State/UT'].unique())

# Counts the frequency of each unique value in a column
# Syntax: df['ColumnName'].value_counts()
print(df['State/UT'].value_counts())

# Shows the index (row labels) of the DataFrame
# Syntax: df.index
print(df.index)

# Converts the column names into a list
# Syntax: df.columns.tolist()
print(df.columns.tolist())

# Displays memory usage of each column including objects
# Syntax: df.memory_usage(deep=True)
print(df.memory_usage(deep=True))
```

```python
# Data Cleaning (3 queries)
# 1. Check for NaN and fill with 0
df.fillna(0, inplace=True)

# 2. Convert all column names to string
df.columns = df.columns.astype(str)

# 3. Strip whitespaces in State/UT column
df['State/UT'] = df['State/UT'].str.strip()
```

```python
# Data Filtering (2 queries)
# States with more than 5000 cases in 2020
df[df['2020'] > 5000]

# States with 0 cases in 2002
df[df['2002'] == 0]
```

```python
# 4. Grouping (2 queries)
# Total cases in 2018-2020 per state
df['Total_2018_2020'] = df[['2018', '2019', '2020']].sum(axis=1)
df.groupby('State/UT')['Total_2018_2020'].sum().sort_values(ascending=False)

# Average cases in early years
df['Avg_2002_2005'] = df[['2002', '2003', '2004', '2005']].mean(axis=1)
df.groupby('State/UT')['Avg_2002_2005'].mean()
```

```python
# 5. Sorting (2 queries)

# Sort by highest 2020 cases
df.sort_values(by='2020', ascending=False).head()

# Sort by least 2002 cases
df.sort_values(by='2002').head()
```

```python
# 6. Aggregation (1 query)
print("Min:", df['2020'].min())
print("Max:", df['2020'].max())
print("Mean:", df['2020'].mean())
print("Mode:", df['2020'].mode()[0])
print("Variance:", df['2020'].var())
```

```python
# 7. Data Visualization (5 Plots)
# Bar plot of top 10 states in 2020
top10_2020 = df.sort_values(by='2020', ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(data=top10_2020, x='State/UT', y='2020')
plt.title('Top 10 States by Cyber Crime in 2020')
plt.xticks(rotation=45)
plt.show()

# Line plot of all years for a state
state = 'MAHARASHTRA'
plt.plot(df.columns[1:-2], df[df['State/UT'] == state].iloc[0,1:-2])
plt.title(f"Year-wise Cases for {state}")
plt.xticks(rotation=90)
plt.ylabel("Cases")
plt.show()

# Heatmap for correlation
plt.figure(figsize=(12, 6))
sns.heatmap(df.iloc[:,1:-2].corr(), cmap='coolwarm', annot=False)
plt.title("Yearly Correlation of Cyber Crimes")
plt.show()

# Boxplot for yearly comparison
plt.figure(figsize=(12, 5))
sns.boxplot(data=df.iloc[:,1:-2])
plt.xticks(rotation=90)
plt.title("Distribution of Cases Across Years")
plt.show()

# Pie chart of total cases share (top 5)
top5 = df[['State/UT', '2020']].sort_values(by='2020', ascending=False).head(5)
plt.pie(top5['2020'], labels=top5['State/UT'], autopct='%1.1f%%', startangle=140)
plt.title("Top 5 States Share in 2020 Cyber Crimes")
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import Normalize
import numpy as np

# Load cybercrime dataset
df = pd.read_csv("cc.csv")
years = [str(y) for y in range(2002, 2021)]
df['State/UT'] = df['State/UT'].str.upper()

# Calculate required metrics per state
df['Average_Crime'] = df[years].mean(axis=1)
df['Min_Crime'] = df[years].min(axis=1)
df['Max_Crime'] = df[years].max(axis=1)
df['Least_Year'] = df[years].idxmin(axis=1)
df['Most_Year'] = df[years].idxmax(axis=1)

# Load map and prepare
india_map = gpd.read_file("india_state.geojson")
india_map['NAME_1'] = india_map['NAME_1'].str.upper()

# Merge with cybercrime data
merged = india_map.merge(df, left_on='NAME_1', right_on='State/UT')

# Create helper dicts for finding min/max state across states
state_avg_dict = dict(zip(df['State/UT'], df['Average_Crime']))
min_state = df.loc[df['Average_Crime'].idxmin(), 'State/UT']
max_state = df.loc[df['Average_Crime'].idxmax(), 'State/UT']

# Plotting
fig, ax = plt.subplots(figsize=(20, 15))
norm = Normalize(vmin=merged['Average_Crime'].min(), vmax=merged['Average_Crime'].max())
merged.plot(column='Average_Crime', cmap='cividis', linewidth=0.8, ax=ax, edgecolor='black', legend=True)

# Add text boxes on each state
for idx, row in merged.iterrows():
    if row['geometry'].centroid.is_empty:
        continue
    x, y = row['geometry'].centroid.x, row['geometry'].centroid.y
    text = (f"{row['State/UT']}\n"
            f"Avg: {int(row['Average_Crime'])}\n"
            f"Min: {int(row['Min_Crime'])} ({row['Least_Year']})\n"
            f"Max: {int(row['Max_Crime'])} ({row['Most_Year']})\n"
            f"Least S: {min_state}\n"
            f"Most S: {max_state}")
    ax.text(x, y, text, fontsize=6, ha='center', va='center', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", lw=0.5))

ax.set_title("Average Cyber Crimes in Indian States (2002–2020) with State-wise Summary", fontsize=18)
ax.axis('off')
plt.tight_layout()
plt.show()
from matplotlib.colors import LightSource
import matplotlib.pyplot as plt
import matplotlib.cm as cm  # Import colormap properly

# Create light source for shaded relief
ls = LightSource(azdeg=315, altdeg=45)
cmap = cm.get_cmap('cividis')  # Properly get the colormap object
shaded_relief = ls.shade(merged['Average_Crime'].values.reshape(merged.shape[0], 1), cmap=cmap, blend_mode='overlay')

# Update plotting with relief shading
fig, ax = plt.subplots(figsize=(20, 15))
merged.plot(column='Average_Crime', cmap='cividis', linewidth=0.8, ax=ax, edgecolor='black', legend=True, alpha=0.9)

# Overlay the shaded relief for depth effect
ax.imshow(shaded_relief, extent=ax.get_xlim() + ax.get_ylim(), origin='upper', alpha=0.5)
ax.set_title("3D Cyber Crimes Map with Topographic Effects", fontsize=18)
ax.axis('off')
plt.show()


```
